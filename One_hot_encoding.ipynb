```python
import pandas as pd
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
#join these 2 tables
df_dummies = pd.concat([df,dummies] , axis = 1)
df_dummies

df_dummies.drop(['town' , 'west windsor'] , axis = 1 , inplace = True)
df_dummies

#dropped one town from any 3
#here from monroe township , west windsor ,robinsville we should delete one as it can be derived
#from other 2 , and if not dropped may cause problem in model training further
```

```python
#extracting the target values
x = df_dummies.drop('price' , axis = 1)
x

#extracting
y = df_dummies.price
y
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)
```

```python
#predicting price of 3400 area in westWindsor
model.predict([[3400 , 0 , 0]])
```

```python
#predicting pirce of 2800 in robinsville
model.predict([[2800 , 0 ,1]])
```

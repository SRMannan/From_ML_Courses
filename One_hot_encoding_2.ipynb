```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
dummies = pd.get_dummies(df.Car)
dummies
```

```python
df_dummies = pd.concat([df , dummies] , axis = 1)
df_dummies
```

```python
df_dummies.drop(['Car' , 'Audi A5'] , axis = 1 , inplace = True)
df_dummies
```

```python
X = df_dummies.drop('Sell Price($)' , axis = 1 )
print(X)

y = df_dummies['Sell Price($)']
print(y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(X,y)
```

```python
model.predict([[45000 , 4 , 0 , 1]])
```

```python
model.predict([[86000 , 7 , 1 , 0]])
```

```python
#printing model accuracy
acc = model.score(X,y)
print(acc * 100 , '%')
```
